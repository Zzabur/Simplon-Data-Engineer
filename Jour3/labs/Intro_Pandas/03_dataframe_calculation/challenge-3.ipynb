We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub Problem 1
# je veux rassembler les CustomerID identiques et additionner leur amount_spent
customerbill = orders.groupby("CustomerID", as_index=False)["amount_spent"].sum()
customerbill = customerbill.sort_values(by="amount_spent", ascending=False)
customerbill.head()
```

```python
# Sub Problem 2
# Je veux diviser les amount_spent en catégories en fonction de leur valeur. Puis faire ressortir les clients en fonction de leur catégories dans amount_spent
VIP = customerbill["amount_spent"].quantile(0.95)
Preferred = customerbill["amount_spent"].quantile(0.75)
print(VIP)
print(Preferred)

clients_VIP = customerbill[customerbill["amount_spent"] > VIP]
print(clients_VIP)
clients_Preferred = customerbill[(Preferred < customerbill["amount_spent"]) & (customerbill["amount_spent"] <= VIP)]
print(clients_Preferred) 
```

```python
# Sub Problem 3
# df['Combo Type'] = df.apply(lambda row: f"{row['Type 1']}-{row['Type 2']}" if pd.notna(row['Type 2']) else row['Type 1'],axis=1)
# def labels(row):
#     if row["CustomerID"] in clients_VIP["CustomerID"].values:
#         return "VIP"
#     elif row["CustomerID"] in clients_Preferred["CustomerID"].values:
#         return "Preferred"
#     else:
#         return "-"
        
# orders["label"] = orders.apply(labels, axis=1)
def labels(row):
    if row["CustomerID"] in clients_VIP["CustomerID"].values:
        return "VIP"
    elif row["CustomerID"] in clients_Preferred["CustomerID"].values:
        return "Preferred"
    else:
        return "-"
orders["label"] = orders.apply(labels, axis=1)
orders.head(10)
```
